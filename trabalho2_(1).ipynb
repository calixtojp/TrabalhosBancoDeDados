# Criação das tabelas

```python
%%sql

DROP TABLE IF EXISTS Localizacao CASCADE;
CREATE TABLE Localizacao (
    idLocalizacao SERIAL PRIMARY KEY,
    cidade        VARCHAR(50)   NOT NULL,
    estado        VARCHAR(50)   NOT NULL,
    pais          VARCHAR(50)   NOT NULL,
    bairro        VARCHAR(50)   NOT NULL,
    UNIQUE  (cidade, estado, pais, bairro)
);

DROP TABLE IF EXISTS ContaBanco CASCADE;
CREATE TABLE ContaBanco (
  nroConta              VARCHAR(20) PRIMARY KEY,
  nroRoteamento         VARCHAR(20) NOT NULL,
  tipoConta             VARCHAR(20) NOT NULL
);

DROP TABLE IF EXISTS Usuario CASCADE;
CREATE TABLE Usuario (
    idUsuario SERIAL PRIMARY KEY,
    nome          VARCHAR(40)     NOT NULL,
    sobrenome     VARCHAR(100)    NOT NULL,
    telefone      VARCHAR(20)     NOT NULL,
    idLocalizacao INTEGER         NOT NULL,
    nroContaBanco VARCHAR(20)     NOT NULL,
    tipoUsuario   CHAR(1)         NOT NULL,
    endereco      VARCHAR(100)    NOT NULL,
    sexo          CHAR(1),
    dataNascimento  DATE,
    email         VARCHAR(40),
    senha         VARCHAR(16),
    UNIQUE  (nome, sobrenome, telefone),
    FOREIGN KEY (idLocalizacao) REFERENCES Localizacao(idLocalizacao),
    FOREIGN KEY (nroContaBanco) REFERENCES ContaBanco(nroConta)
);

DROP TABLE IF EXISTS Propriedade CASCADE;
CREATE TABLE Propriedade (
  idPropriedade            SERIAL PRIMARY KEY,
  nome          VARCHAR(100)  NOT NULL,
  endereco      VARCHAR(100)    NOT NULL,
  classe        VARCHAR(50)   NOT NULL,
  pernoite      DECIMAL(10, 2)NOT NULL,
  nroMaxHospedes  INTEGER       NOT NULL,
  idLocalizacao INTEGER         NOT NULL,
  nroQuartos     DECIMAL(4)    NOT NULL,
  nroBanheiros   INTEGER       NOT NULL,
  nroMinNoites  INTEGER       NOT NULL,
  nroMaxNoites  INTEGER       NOT NULL,
  valorTaxLimpeza  DECIMAL(10, 2) NOT NULL,
    UNIQUE  (nome, endereco),
  FOREIGN KEY (idLocalizacao) REFERENCES Localizacao(idLocalizacao)
);

DROP TABLE IF EXISTS Oferta CASCADE;
CREATE TABLE Oferta (
    idOferta SERIAL PRIMARY KEY,
    dataIni        DATE   NOT NULL,
    idPropriedade        INTEGER   NOT NULL,
    idUsuario          INTEGER   NOT NULL,
    dataFim        DATE   NOT NULL,
    valorDesconto DECIMAL(10, 2) NOT NULL,
    valorTotal DECIMAL(10, 2) NOT NULL,
    nroHospedes INTEGER          NOT NULL,
    impostoPago DECIMAL(10, 2)      NOT NULL,
    taxaLimpeza DECIMAL(10, 2) NOT NULL,
    codPromo VARCHAR(20),
    UNIQUE  (dataIni, idPropriedade, idUsuario)
);

DROP TABLE IF EXISTS Locacao CASCADE;
CREATE TABLE Locacao (
    idLocacao SERIAL PRIMARY KEY,
    checkIn        DATE   NOT NULL,
    idOferta        INTEGER   NOT NULL,
    idUsuario          INTEGER   NOT NULL,
    checkOut        DATE   NOT NULL,
    dataReserva     DATE NOT NULL,
    confirmadaTF CHAR(1),
    UNIQUE  (checkIn, idOferta, idUsuario),
    FOREIGN KEY (idOferta) REFERENCES Oferta(idOferta),
    FOREIGN KEY (idUsuario) REFERENCES Usuario(idUsuario)
);

DROP TABLE IF EXISTS Mensagem CASCADE;
CREATE TABLE Mensagem (
  idMensagem SERIAL PRIMARY KEY,
  idUserOrigem INTEGER,
  idUserDestino INTEGER,
  texto TEXT NOT NULL,
  idLocação INTEGER NOT NULL,
  timeStamp TIMESTAMP NOT NULL,
  UNIQUE  (idUserOrigem, idUserDestino, timeStamp),
  FOREIGN KEY (idUserOrigem) REFERENCES Usuario(idUsuario),
  FOREIGN KEY (idUserDestino) REFERENCES Usuario(idUsuario)
);

DROP TABLE IF EXISTS Avaliacao CASCADE;
CREATE TABLE Avaliacao (
  idAvaliacao SERIAL PRIMARY KEY,
  idUsuario INTEGER,
  idLocacao INTEGER,
  mensagem TEXT,
  notaLimpeza DECIMAL(1)     NOT NULL,
  notaComunicação DECIMAL(1) NOT NULL,
  notaLocação DECIMAL(1) NOT NULL,
  notaValor DECIMAL(1)       NOT NULL,
  UNIQUE  (idUsuario, idLocacao),
  FOREIGN KEY (idLocacao) REFERENCES Locacao(idLocacao),
  FOREIGN KEY (idUsuario) REFERENCES Usuario(idUsuario)
);

DROP TABLE IF EXISTS Regra CASCADE;
CREATE TABLE Regra (
  idRegra         SERIAL PRIMARY KEY,
  idPropriedade   INTEGER,
  descricao       TEXT NOT NULL,
  FOREIGN KEY (idPropriedade) REFERENCES Propriedade(idPropriedade)
);

DROP TABLE IF EXISTS Comodidade CASCADE;
CREATE TABLE Comodidade (
  idComodidade    SERIAL PRIMARY KEY,
  idPropriedade   INTEGER,
  descricao       TEXT NOT NULL,
  FOREIGN KEY (idPropriedade) REFERENCES Propriedade(idPropriedade)
);

DROP TABLE IF EXISTS PontoInteresse CASCADE;
CREATE TABLE PontoInteresse (
  idPontoInteresse    SERIAL PRIMARY KEY,
  idLocalizacao   INTEGER,
  descricao       TEXT NOT NULL,
  FOREIGN KEY (idLocalizacao) REFERENCES Localizacao(idLocalizacao)
);

DROP TABLE IF EXISTS Foto CASCADE;
CREATE TABLE Foto (
  idFoto SERIAL PRIMARY KEY,
  idAvaliacao INTEGER,
  path VARCHAR(255) NOT NULL,
  FOREIGN KEY (idAvaliacao) REFERENCES Avaliacao(idAvaliacao)
);

DROP TABLE IF EXISTS Quarto CASCADE;
CREATE TABLE Quarto (
  idQuarto        SERIAL PRIMARY KEY,
  idPropriedade   INTEGER           NOT NULL,
  compartilhadoTF CHAR(1) NOT NULL,
  nroCamas   DECIMAL(2)        NOT NULL,
  tipoCama    VARCHAR(30) NOT NULL,
  FOREIGN KEY (idPropriedade) REFERENCES Propriedade(idPropriedade)
);
```

## Exercício 7

Crie uma visão chamada PropriedadesDisponíveis que mostra as propriedades disponíveis para a locação em 2024.

### 7.1
• Crie o comando da visão.

```python
%%sql
CREATE VIEW PropriedadesDisponiveis AS
    SELECT
        propriedades.*
    FROM
        oferta
    JOIN propriedades ON
        (oferta.idpropriedade = propriedades.idpropriedade)
    WHERE
        EXTRACT(YEAR FROM dataini) = 2024
    OR EXTRACT(YEAR FROM datafim) = 2024
    GROUP BY propriedades.idpropriedade, propriedades.nome, propriedades.classe,
            propriedades.pernoite, propriedades.nrominnoites, propriedades.idlocalizacao
```

### 7.2
• Execute uma consulta usando essa visão que mostre preço por noite médio de todas as propriedades disponíveis para a locação em cada um dos meses de 2024.

```python
%%sql
SELECT
    EXTRACT(MONTH FROM o.dataini) AS mes,
    AVG(p.pernoite) AS preco_medio_por_noite
FROM
    PropriedadesDisponiveis p
JOIN oferta o ON
    p.idpropriedade = o.idpropriedade
WHERE
    EXTRACT(YEAR FROM o.dataini) = 2024
GROUP BY
    EXTRACT(MONTH FROM o.dataini)
ORDER BY
    mes;
```

## Exercício 8

### 8.1
• Mostre os anfitriões que têm pelo menos 3 imóveis, trazendo seu nome, sua cidade e a quantidade de imóveis dos quais ele é dono.

```python
%%sql
SELECT
    o.nome AS nome_anfitriao,
    localizacao.cidade,
    COUNT(o.idpropriedade) AS qtd_imoveis
FROM
    (SELECT
        oferta.idpropriedade, oferta.idusuario, usuario.nome
    FROM
        usuario
    JOIN oferta ON
        usuario.idusuario = oferta.idusuario
    WHERE
        usuario.anfitriaotf = 'T'
    GROUP BY
        oferta.idpropriedade, oferta.idusuario, usuario.nome) as o
JOIN propriedades ON
    o.idpropriedade = propriedades.idpropriedade
JOIN localizacao ON
    localizacao.idlocalizacao = propriedades.idlocalizacao
GROUP BY
    o.nome,
    localizacao.cidade
HAVING
    COUNT(o.idpropriedade) >= 3
```
